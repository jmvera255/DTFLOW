# Dimension Reduction via diffusion propagation and Bhattacharyya Kernel

The key innovation is the combination of the **random walk with restart (RWR) method** and **Bhattacharyya kernel matrix**, 
which forms a new manifold learning method for discovering low-dimensional data structure and approximate **geodesic distance** calculation.

### Main Reference

【1】. L. Haghverdi, M. Buettner, F. A. Wolf, F. Buettner, F. J. Theis, Diffusion pseudotime robustly reconstructs lineage branching, Nature methods 13 (10) (2016) 845–848.

【2】. S. Wang, H. Cho, C. Zhai, B. Berger, J. Peng, Exploiting ontology graph for predicting sparsely annotated gene function, Bioinformatics 31 (12) (2015) i357–i364.

```python
import numpy as np
import pandas as pd

from scipy import spatial,linalg,sparse,stats
from sklearn import datasets,decomposition

import matplotlib.pyplot as plt
import matplotlib.cm as cm
# import seaborn as sns
```

### Step1. Markov Adjacency Matrix Construction
define Gaussian kernel:
$$
K(\boldsymbol{x_i},\boldsymbol{x_j}) =\exp\left(-\frac{||\boldsymbol{x_i}-\boldsymbol{x_j}||^2}{2\sigma_i\sigma_j}\right), \label{eq-gauss}\tag{1}
$$
normalize the kernel:
$$
D(\boldsymbol{x_i}) = \sum_j {K}(\boldsymbol{x_i},\boldsymbol{x_j})
$$

$$
M_{ij} = \frac{{K}(\boldsymbol{x_i},\boldsymbol{x_j})}{D(\boldsymbol{x_i})}, \label{eq-m} 
$$

```python
def get_transition_matrix(X,scale =1):
    kdtree = spatial.cKDTree(X)
    # get kernel width
    r0 = np.max([kdtree.query(X[i,:],k=2)[0][1] for i in range(X.shape[0])])
    epsilon = r0 * scale 
    # get distance matrix
    coo_dist_matrix = spatial.cKDTree.sparse_distance_matrix(kdtree,kdtree,epsilon,output_type='coo_matrix') 
    # get weight matrix
    transformed_data = np.exp(-np.square(coo_dist_matrix.data) / (2.0 * epsilon *epsilon)) 
    kernel_matrix =  coo_dist_matrix._with_data(transformed_data, copy=True)
    row_sums = np.asarray(kernel_matrix.sum(axis=1)).squeeze()
    # alpha = 0.5
    # D_sum = np.power(row_sums,-alpha).flatten()
    # M = np.diag(D_sum)@ kernel_matrix @ np.diag(D_sum)
    # mevals, mevecs =  np.linalg.eigh(M)
    M = kernel_matrix @ np.diag(np.power(row_sums,-1).flatten())
    return M.T 
```

### Step2. Random walk with restart

Random walk with restart satisfies the following equation:

$$\boldsymbol{s_i}^{k}=p\boldsymbol{s_i}^{k-1}M+(1-p)\boldsymbol{e_i}, \quad 0<p<1 \label{eq-rwr}$$

$$\boldsymbol{s_i} = \boldsymbol{s_i}^\infty = (1-p)\boldsymbol {e_i}(I-pM)^{-1}, \label{eq-s}$$

the diffusion matrix:

$$S =(1-p)(I-pM)^{-1}. \label{eq-ss}$$
$$\forall s_{ij}\in S, s_{ij}>0,\;\text{and}\;\sum_{j=1}^N s_{ij} = 1$$

```python
def diffusionRWR(M,prob=0.8):
    I = np.identity(M.shape[0])
    S = (1-prob) * np.linalg.inv(I-prob*M)
    return S
```

### Step3. Learning embedding

Bhattacharyya coefficient(BC):
$$BC (\boldsymbol{p} , \boldsymbol{q}) = \sum_{i = 1}^n \sqrt {p_{i} q_{i}}$$

Bhattacharyya kernel matrix:

$$G = {\sqrt{S} } \sqrt{S}^T= \left[\langle\sqrt{\boldsymbol{s_i}},\sqrt{\boldsymbol{s_j}}\rangle_{i,j=1,\cdots,N}\right]_{N\times N}$$

According to the Mercer’s Theorem, there must exist one kernel function $k$:

$$k(\boldsymbol{s_i},\boldsymbol{s_j}) = G_{ij} = \langle\sqrt{\boldsymbol{s_i}},\sqrt{\boldsymbol{s_j}}\rangle,\forall i,j\in[1,\cdots,N].$$

$$G_{ij}=k(\boldsymbol{s_i},\boldsymbol{s_j})\triangleq e^{{k_1(\boldsymbol{s_i},\boldsymbol{s_j})}}=e^{\langle\boldsymbol\phi(\boldsymbol{s_i}),\boldsymbol\phi(\boldsymbol{s_j})\rangle}=e^{\langle\boldsymbol{y_i},\boldsymbol{y_j}\rangle}.
\label{eq-bk}$$
here we let $\boldsymbol{y_i} = \boldsymbol\phi(\boldsymbol{s_i})$, then
$$\langle\boldsymbol{y_i},\boldsymbol{y_j}\rangle = \log G_{ij}$$

$$\log G = Y^T Y = V\Sigma V^T,\label{eq-lt}$$

##### Single Value Decomposition

$$Y_1 = Y_d = V_d\Sigma_d^{1/2}$$

##### Multidimensional Scaling
since $||\boldsymbol{y_i}||^2=\log G_{ii}=0\label{eq-sdist}$,
$$D^{(2)} = ||\boldsymbol{y_i}-\boldsymbol{y_j}||^2=||\boldsymbol{y_i}||^2+||\boldsymbol{y_j}||^2-2\langle\boldsymbol{y_i},\boldsymbol{y_j}\rangle = -2\log G_{ij},\tag{2}$$

Notice that Exponent (1) and Logarithm (2) are inverse operations. 
$$B=-\frac{1}{2}JD^{(2)}J,J = I_N-\frac{1}{N}\mathbf{1}\mathbf{1}^T$$
$$B=V\Lambda V^T$$
$$Y_2 =Y_d = V_d\Sigma_d^{1/2}$$

###### Laplacian Eigenmaps

$$D_G=diag(\sum_{j=1}^N G_{ij})$$
$$L_G = D_G-G$$
$$L_G\boldsymbol{y}=\lambda D_G\boldsymbol{y}$$

```python
def learnVector(S,dim=2):
    
    N = S.shape[0]
    A = np.sqrt(S)
    G = A @ A.T
    H = np.log(G)
    
    # Single Value Decomposition
    #  U, s, V = sparse.linalg.svds(H, k=dim)
    #  Sig = s * np.identity(dim)
    #  Y1 = U @ np.sqrt(Sig)
    Y1 = decomposition.TruncatedSVD().fit_transform(H) 

    # Multidimensional Scaling
    # H = 0.5 * np.add.outer(np.diag(H),np.diag(H))- H
    # np.fill_diagonal(H,0) 
    J = np.eye(N) - np.ones((N, N))/N
    B = J @ H @ J
    Y2 = decomposition.TruncatedSVD().fit_transform(B) 

    # Laplacian Eigenmaps
    Dg = np.diag(np.sum(G,1))
    Lg = Dg - G
    eval, evec  = linalg.eigh(Lg,Dg)
    Y3 = evec[:,1:dim+1]

    return Y1,Y2,Y3
```

### Example

```python
digits= datasets.load_digits()
X = np.vstack([digits.data[digits.target==i] for i in range(10)])
y = np.hstack([digits.target[digits.target==i] for i in range(10)])
```

```python
M = get_transition_matrix(X,scale=1)
S = diffusionRWR(M,prob=0.99)
Y1,Y2,Y3 = learnVector(S)
```

**Interesting phenomena: the structure of Multidimensional Scaling(B) and Laplacian Eigenmaps(C) is very similar.**

**The distance between digit 2 and 4 is close in Figure A. However the distance between 2 and 4 is far in Figure B and C.**

```python
fig, axes = plt.subplots(nrows=1,ncols=3,figsize=(40,15))
color_stage = sorted(set(y))

ax1 = plt.subplot(131)
ax1.scatter(Y1[:,1],Y1[:,0], c=y,s=10,cmap=cm.jet)
for i in color_stage:
    indices = np.where(y == i)[0]
    mn = np.mean(Y1[indices,:],axis=0)
    ax1.scatter(mn[1],mn[0],s=10)
    ax1.annotate(i,xy=(mn[1],mn[0]),fontsize=50)

ax2 = plt.subplot(132)
ax2.scatter(Y2[:,0],Y2[:,1], c=y,s=10,cmap=cm.jet)
for i in color_stage:
    indices = np.where(y == i)[0]
    mn = np.mean(Y2[indices,:],axis=0)
    ax2.scatter(mn[0],mn[1],s=10)
    ax2.annotate(i,xy=(mn[0],mn[1]),fontsize=50)

ax3 = plt.subplot(133)
ax3.scatter(-Y3[:,0],-Y3[:,1], c=y,s=10,cmap=cm.jet)
ax3.set_xlim((-0.0015,0.0015))
ax3.set_ylim((-0.001,0.0015))
for i in color_stage:
    indices = np.where(y == i)[0]
    mn = np.mean(Y3[indices,:],axis=0)
    ax3.scatter(-mn[0],-mn[1],s=10)
    ax3.annotate(i,xy=(-mn[0],-mn[1]),fontsize=50)

def get_axis_limits(ax,a,b):
    return ax.get_xlim()[1]*a, ax.get_ylim()[1]*b
ax1.annotate('A', xy=get_axis_limits(ax1,0.89, 0.975),fontsize=35, fontweight='bold')
ax2.annotate('B', xy=get_axis_limits(ax2,0.89, 0.93),fontsize=35, fontweight='bold')
ax3.annotate('C', xy=get_axis_limits(ax3,0.89, 0.93),fontsize=35, fontweight='bold')
```
